```python
import pandas as pd
```

```python
def read_markup(dataset_name):
    markup = pd.read_csv(f"data/markup_{dataset_name}.tsv", sep="\t")
    markup.columns = markup.columns.map(lambda x: x.lower().replace(":", "_"))
    return markup
```

```python
# чтение файла разметки манипуляции из пула https://sandbox.toloka.yandex.ru/requester/project/83394/pool/1046375
markup = read_markup("first_pool")
markup.head(2)
```
